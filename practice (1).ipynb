```python
numbers = [12, 7, 23, 48, 15, 66, 9]

sum_even = 0

for num in numbers:
    if num % 2 == 0:
        sum_even += num

print("Sum of even numbers =", sum_even)
```

```python
data = ['apple', 'banana', 'cherry', 'date', 'elderberry']

reverse_data = []

for i in range(len(data)-1, -1, -1):
    reverse_data.append(data[i])

print(reverse_data)
```

```python
values = [8, 2, 10, 5, 17, 3, 14]

largest = values[0]

for num in values:
    if num > largest:
        largest = num

print("Largest number:", largest)
```

```python
items = ['red', 'blue', 'green', 'red', 'yellow', 'blue', 'red']

count = 0

for color in items:
    if color == 'red':
        count += 1

print("Count of red:", count)
```

```python
mixed_numbers = [1, 5, 2, 5, 3, 5, 4, 5]

new_list = []

for num in mixed_numbers:
    if num != 5:
        new_list.append(num)

print("Final list:", new_list)
```
