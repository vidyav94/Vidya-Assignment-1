```python
""" 
use a loop and if-else to check if a specific character is present in the string or not.
get output for both if the character is present and if it is not present.
take 2 characters to check in the string 

user input -> string, character1, character2
""";
```

```python
text = input("Enter a string: ")

char1 = input("Enter the first character to check: ")
char2 = input("Enter the second character to check: ")

found_char1 = False
found_char2 = False

for char in text:
    if char == char1:
        found_char1 = True
    if char == char2:
        found_char2 = True

if found_char1:
    print(f"The character '{char1}' is present in the string.")
else:
    print(f"The character '{char1}' is not present in the string.")

if found_char2:
    print(f"The character '{char2}' is present in the string.")
else:
    print(f"The character '{char2}' is not present in the string.")
```
