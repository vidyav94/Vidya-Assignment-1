```python
# function to check which number is larger among two numbers
```

```python
def compare(a,b):
    if a>b:
        print(a,"is larger")
    elif b>a:
        print(b,"is larger")
    else:
        print("both are equal")
```

```python
# create a function to find the factorial of a number
```

```python
def fact(a):
    f = 1
    for i in range(1,a+1):
        f = f*i
    print(f)
```

```python
fact(5)
```

```python
fact(10)
```

```python
def ini(*args):
    return args
```

```python
ini(10, 20, 30, 40, 5, 1, 5)
```

```python
# pass any number of integers in a function and find their sum
```

```python
def addall(*args):
    s = 0
    for i in args:
        s = s + i
    return s
```

```python
addall(10, 20, 30, 40, 5, 1, 5)
```

```python
# type hinting => tell the type of parameters and return value, and docstring
```

```python
def sq(x: int) -> int:

    """ 
    Calculate the square of a number.
    
    Args:
        x (int): The number to square.
        
    Returns:
        int: The square of the number.
    """

    res = x*x
    return res
```

```python
help(sq)
```
