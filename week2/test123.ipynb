```python
print("welcome")
text = input("enter the string")

print("menu =>\n1. upper\n2. lower\n3. replace")

ch = int(input("enter the choice"))

if ch == 1:
    print(f"{text.upper()}")
elif ch == 2:
    print(f"{text.lower()}")
elif ch == 3:
    r1 = input("which part u want to replace: ")
    r2 = input("what u want to replace it with")
    new = text.replace(r1, r2)
    print(f"result -> {new}")
```
