```python
'''1. Write a function safe_divide(a, b) that divides two numbers and handles ZeroDivisionError and
TypeError separately, printing an appropriate message for each. Use a finally block to print
'Operation attempted.' regardless of the outcome. Test it with valid numbers, division by zero, and a
string input.''';
```

```python
def divide(a,b):
    try:
        if a/b:
            print("Result=",a/b)
    except ZeroDivisionError:
        print("No can't divisible by zero")
    except TypeError:
        print("Invalid datatype")
    finally:
        print("Code executed successfully")
```

```python
'''2. Write a function validate_age(age) that raises a custom exception InvalidAgeError (a class you
define, inheriting from Exception) if age is negative or greater than 120. Catch this exception in a
calling block and print a user-friendly message. Test it with at least 3 different values.''';
```

```python
class InvalidAgeError(Exception):
    pass

def validate_age(age):
    if age < 0 or age > 120:
        raise InvalidAgeError("Age must be between 0 and 120")
    print("Valid Age:", age)

ages = [25, -10, 130]

for age in ages:
    try:
        validate_age(age)
    except InvalidAgeError as e:
        print("Invalid Age Error:", e)
```

```python
'''3. Write a program that repeatedly asks the user (use a loop with at least 3 simulated inputs in a list,
e.g. user_inputs = ['10', 'abc', '0', '5']) to enter a number and divide 100 by it. Handle the following in
a single try block using multiple except clauses:
a) ValueError if the input is not a number.
b) ZeroDivisionError if the input is zero.
c) A generic Exception as a catch-all for anything else.
Print the result if successful, otherwise print which error occurred.''';
```

```python
values = ["10", "abc", "0", "5"]

for item in values:
    try:
        number = int(item)
        answer = 100 / number
        print("Answer =", answer)

    except ValueError:
        print("ValueError: Not a number")

    except ZeroDivisionError:
        print("ZeroDivisionError: Cannot divide by zero")

    except Exception as e:
        print("Other Error:", e)
```

```python
'''7. Create a Pandas DataFrame from the dictionary below and perform the following:
data = { "Name": ["Aarav", "Priya", "Karan", "Meera", "Sahil"], "Age": [23, 21,
25, 22, 24], "Department": ["AI", "Data Science", "AI", "Web Dev", "Data
Science"], "Score": [88, 92, 79, 95, 67], }
a) Display the first 3 rows and the data types of each column.
b) Add a new column Grade based on Score (>=90 'A', >=75 'B', else 'C').
c) Sort the DataFrame by Score in descending order.
d) Filter rows where Department is 'AI'.''';
```

```python
import pandas as pd

data = {
    "Name": ["Aarav", "Priya", "Karan", "Meera", "Sahil"],
    "Age": [23, 21, 25, 22, 24],
    "Department": ["AI", "Data Science", "AI", "Web Dev", "Data Science"],
    "Score": [88, 92, 79, 95, 67]
}

df = pd.DataFrame(data)

# a) First 3 rows and data types
print("First 3 Rows:")
print(df.head(3))

print("\nData Types:")
print(df.dtypes)

# b) Add Grade column
grade = []

for i in df["Score"]:
    if i >= 90:
        grade.append("A")
    elif i >= 75:
        grade.append("B")
    else:
        grade.append("C")

df["Grade"] = grade

print("\nDataFrame with Grade:")
print(df)

# c) Sort by Score in descending order
sorted_df = df.sort_values(by="Score", ascending=False)

print("\nSorted DataFrame:")
print(sorted_df)

# d) Filter Department = AI
ai_df = df[df["Department"] == "AI"]

print("\nAI Department Students:")
print(ai_df)
```

```python
import pandas as pd

data = {
    "Name": ["Aarav", "Priya", "Karan", "Meera", "Sahil"],
    "Age": [23, 21, 25, 22, 24],
    "Department": ["AI", "Data Science", "AI", "Web Dev", "Data Science"],
    "Score": [88, 92, 79, 95, 67]
}

df = pd.DataFrame(data)

avg_score = df.groupby("Department")["Score"].mean()
print("Average Score for each",avg_score)





index = df["Score"].idxmax()
print(f"The row with the highest Score{ df.loc[index]}")





df = df.rename(columns={"Score": "Marks"})
print(df)

df = df.drop("Age", axis=1)
print(df)
```

```python
'''9. Create a DataFrame with some missing values as shown below and handle them:
import numpy as np data = { "Product": ["Pen", "Notebook", "Eraser",
"Sharpener", "Ruler"], "Price": [10, np.nan, 5, 8, np.nan], "Stock": [100, 50,
np.nan, 30, 20], }
a) Check the number of missing values in each column using isnull().sum().
b) Fill missing Price values with the column mean.
c) Drop rows where Stock is still missing.
d) Save the cleaned DataFrame to a CSV file named cleaned_products.csv without the index''';
```

```python
import numpy as pd
import pandas as pd
data=data = { "Product": ["Pen", "Notebook", "Eraser",
"Sharpener", "Ruler"], "Price": [10, np.nan, 5, 8, np.nan], "Stock": [100, 50,
np.nan, 30, 20],}
df=pd.DataFrame(data)

print(df.isnull().sum())

mean_price = df["Price"].mean()
df["Price"] = df["Price"].fillna(mean_price)

df = df.dropna(subset=["Stock"])

df.to_csv("cleaned_products.csv", index=False)

print(df)
```
