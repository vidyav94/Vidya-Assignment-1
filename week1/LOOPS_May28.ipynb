```python
# write a program to find factorial of a number using while loop

num = int(input("enter the number"))

fact = 1

while num > 0:
    fact = fact * num
    num -= 1

print(fact)
```

```python
num = 12345

add = 0

while num > 0:
    rem = num % 10
    add = add + rem
    num = num // 10

print(add)
```

```python
num = int(input("enter the number"))

fact = 1

for n in range(1, num + 1):
    fact = fact * n

print(fact)
```

```python
for i in range(1, 11):
    if i == 5:
        break
    print(i)
    print("ok")
```
