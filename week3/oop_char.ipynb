```python
class Student:

    def __init__ (self, name, age):
        self.name = name
        self.age = age
        self.skills = []

    def add_skill(self, skill):
        return self.skills.append(skill)
```

## inheritance

```python
class Person:
    insti = "MITU"

    def __init__ (self, age, ht):
        self.age = age
        self.ht = ht

    def show(self):
        print(f"age -> {self.age}, height -> {self.ht}, institute -> {self.insti}")
```

```python
class Ravi(Person):

    def __init__(self, age, ht, sal):
        super().__init__(age, ht)
        self.sal = sal

    def show_sal(self):
        print(f"salary is {self.sal}")
```

```python
a = Ravi(25, 5.9, 57000)
```

```python
a.show_sal()
```

```python
a.show()
```

```python
# add items, view, exit (grocery list)
```

```python
class Menuapp:

    def __init__ (self):
        self.itemlist = []

    def show_menu(self):

        print("Menu\n1.add items\n2. view items\n3. exit")

    def add_items(self):
        item = input("enter the item name:")
        self.itemlist.append(item)

    def view_items(self):
        return self.itemlist
    
    def run(self):

        while True:
            
            self.show_menu()

            ch = int(input("enter the choice"))

            if ch == 1:
                self.add_items()
            elif ch == 2:
                self.view_items()
            elif ch == 3:
                break
            else:
                return 'invalid choice'

```

```python
atharv = Menuapp()
```

```python
atharv.run()
```

```python
atharv.itemlist
```

```python
class Person:

    def __init__ (self, age, name):

        self.__age = age
        self.__name = name

    # getter
    def get_name(self):
        return self.__name
    
    def get_age(self):
        return self.__age
    
    # setter
    def set_age(self, age):
        self.__age = age
        

    def __str__(self):
        return f"age -> {self.__age}, name -> {self.__name}"
```

```python
p = Person(34, 'Arun')
```

```python
print(p)
```
